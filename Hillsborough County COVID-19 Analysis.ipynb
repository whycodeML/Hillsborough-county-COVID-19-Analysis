```python
import numpy as np
import pandas as pd 
```

```python
from datetime import date,datetime
times = pd.date_range(end=date.today().strftime("%m-%d-%y"), periods=30)
```

```python
df_times = pd.DataFrame(times)
df_times.head()
```

```python
# Generating Last 30 days same as the name of the csv file present in the github

import datetime
df_times=[]
for i in times:
    formatted_date = datetime.date.strftime(i, "%m-%d-%Y")
    df_times.append(formatted_date)
print(df_times)
```

```python
df_list = [pd.DataFrame() for df in df_times]
df_dict = dict(zip(df_times, df_list))
```

```python
df_dict.keys()
```

```python
url_part = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/'
```

```python
#Filtering the data for Hillsborough during read, so that we dont pull extra data 

df_final = pd.DataFrame()
Date = []
for i, (key, value) in enumerate(df_dict.items()):
    try:
        value = pd.read_csv(url_part+key+'.csv', iterator=True,error_bad_lines=False)
        #print('i ',i,'key ',key)        
        value = pd.concat([chunk[chunk['Combined_Key'] == 'Hillsborough, Florida, US'] for chunk in value])
        df_final = df_final.append(value)
        Date.append(key)
    except Exception:
        #print('Another Error!!', key+'.csv')
        continue
```

```python
#Adding file date to the dataframe to keep track of the record pulled

df_final['FileDate'] = Date
df_final.head()
```

```python
df_final.reset_index(inplace=True)
```

```python
df_final.drop('index', axis=1, inplace=True)
df_final.head()
```

```python
df_final.to_excel('COVID-19 Daily Data.xls',index=False, sheet_name='Hillsborough')
```
